**Program 5:**  perform exploratory data analysis using Seaborn and Matplotlib libraries. Create differrent types of plots and visualizations to understand data distribution and relationships. use a publicly available dataset

Load the IRIS dataset and print the head of the dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Display the first few rows of the dataset
print(iris.head(100))
```

Pairplot: A pairplot allows us to visualize pairwise relationships in the dataset

```python
sns.pairplot(iris, hue="species")
plt.show()
```

**Boxplot:** Boxplots can be used to visualize the distribution of a single continuous variable or the distribution of a continuous variable within different categories.

```python
sns.boxplot(x="species", y="petal_length", data=iris)
plt.show()
```

**Histogram:** Histograms are useful for visualizing the distribution of a single continuous variable.

```python
sns.histplot(iris["petal_length"], bins=20, kde=True)
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length")
plt.show()
```

**Violin plot:** Violin plots are similar to boxplots but also display the probability density of the data at different values.

```python
sns.violinplot(x="species", y="petal_width", data=iris)
plt.show()
```

**Correlation Heatmap:** A correlation heatmap can be useful to visualize the correlation between different variables in the dataset.

```python
correlation_matrix = iris.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="cool")
plt.title("Correlation Heatmap")
plt.show()
```

**Barplot:** Barplots are helpful for visualizing the distribution of a categorical variable or the relationship between a categorical variable and a continuous variable.

```python
sns.barplot(x="species", y="sepal_width", data=iris)
plt.xlabel("Species")
plt.ylabel("Sepal Width")
plt.title("Mean Sepal Width by Species")
plt.show()
```

**Scatter plot:** Scatter plots are useful for visualizing the relationship between two continuous variables.

```python
sns.color_palette("Spectral", as_cmap=True)
```

```python
sns.scatterplot(x="petal_length", y="petal_width", data=iris, hue="species")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width")
plt.show()
```

**Jointplot:** Jointplots can be used to visualize the relationship between two variables along with their individual distributions.

```python
sns.jointplot(x="sepal_length", y="sepal_width",palette=None, data=iris, kind="scatter")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title(" Jointplot of Sepal Length and Sepal Width")
plt.show()
```

**PairGrid:** PairGrid allows you to create a grid of subplots for pairwise relationships, and it offers flexibility in customizing each subplot.

```python
g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

**Strip plot:** Strip plots are similar to scatter plots but are especially useful for visualizing the distribution of a continuous variable across different categories.

```python
sns.stripplot(x="species", y="petal_length", data=iris, jitter=True)
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.title("Strip Plot of Petal Length by Species")
plt.show()
```

**Swarm plot:** Swarm plots combine elements of strip plots and violin plots to show each data point while avoiding overlap.

```python
sns.swarmplot(x="species", y="petal_width", data=iris)
plt.xlabel("Species")
plt.ylabel("Petal Width")
plt.title("Swarm Plot of Petal Width by Species")
plt.show()
```

**FacetGrid:** FacetGrid allows you to create multiple plots (facets) based on one or more categorical variables.

```python
g = sns.FacetGrid(iris, col="species")
g.map(sns.histplot, "sepal_length")
plt.show()
```

**Boxenplot:** Boxen plots (also known as letter-value plots) are similar to boxplots but provide more information about the tails of the distribution.

```python
sns.boxenplot(x="species", y="sepal_width", data=iris)
plt.xlabel("Species")
plt.ylabel("Sepal Width")
plt.title("Boxen Plot of Sepal Width by Species")
plt.show()
```

**Pairwise Relationships with Kernel Density Estimation (KDE):** PairGrid with KDE plots provides a smoothed estimation of the joint distribution of variables.

```python
g = sns.PairGrid(iris)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot, kde_kws={'color': 'r'})
plt.show()
```
