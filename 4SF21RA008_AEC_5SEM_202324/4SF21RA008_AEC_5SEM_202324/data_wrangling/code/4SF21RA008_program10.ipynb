```python
import plotly.express as px

# using the dataset
df = px.data.tips()

# plotting the scatter chart
fig = px.scatter(df, x='total_bill', y="tip", color='time',
				symbol='day', size='size', facet_row='smoker',
				facet_col='time')

# showing the plot
fig.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def projectile_motion(angle_deg, initial_velocity, gravity=9.81, time_step=0.01):
    angle_rad = np.deg2rad(angle_deg)
    vx = initial_velocity * np.cos(angle_rad)
    vy = initial_velocity * np.sin(angle_rad)
    t = 0
    x_list, y_list = [], []
    while True:
        x = vx * t
        y = vy * t - 0.5 * gravity * t**2
        if y < 0:
            break
        x_list.append(x)
        y_list.append(y)
        t += time_step
    return np.array(x_list), np.array(y_list)

def plot_trajectory(x, y):
    plt.plot(x, y)
    plt.xlabel('Distance (m)')
    plt.ylabel('Height (m)')
    plt.title('Projectile Motion Trajectory')
    plt.grid(True)
    plt.show()

def calculate_max_height(y):
    return np.max(y)

def calculate_range(x):
    return np.max(x)

# Define parameters
initial_velocity = 30  # m/s
launch_angle = 45  # degrees

# Calculate trajectory
x, y = projectile_motion(launch_angle, initial_velocity)

# Generate dataset
data = pd.DataFrame({'Time': np.arange(0, len(x)), 'X': x, 'Y': y})

# Visualize trajectory
plot_trajectory(x, y)

# Calculate and plot maximum height
max_height = calculate_max_height(y)
print("Maximum height reached by the projectile:", max_height, "m")

# Determine the range
range_projectile = calculate_range(x)
print("Range of the projectile:", range_projectile, "m")
```
